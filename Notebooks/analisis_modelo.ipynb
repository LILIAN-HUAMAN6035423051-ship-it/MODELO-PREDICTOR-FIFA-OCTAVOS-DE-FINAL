# Análisis del modelo de predicción de clasificación

## Copa Mundial 2026

Este notebook documenta el proceso experimental utilizado para comparar modelos de Machine Learning orientados a la predicción de equipos clasificados en partidos eliminatorios.

El análisis utiliza información histórica de partidos internacionales y una validación temporal independiente mediante el Mundial 2022.

## 1. Objetivo del proyecto

El objetivo es desarrollar y comparar diferentes algoritmos de clasificación para determinar qué modelo presenta el mejor rendimiento predictivo.

Los modelos evaluados son:

- Regresión Logística
- Árbol de Decisión
- Random Forest
- K-Nearest Neighbors (KNN)
- Support Vector Machine (SVM)
- XGBoost

El modelo seleccionado posteriormente se utiliza para generar predicciones probabilísticas de partidos pendientes de los octavos de final del Mundial 2026.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

RUTA_RESULTADOS = "../Results"

comparacion_modelos = pd.read_csv(
    os.path.join(
        RUTA_RESULTADOS,
        "comparacion_modelos.csv"
    )
)

validacion_2022 = pd.read_csv(
    os.path.join(
        RUTA_RESULTADOS,
        "validacion_mundial_2022.csv"
    )
)

predicciones_2026 = pd.read_csv(
    os.path.join(
        RUTA_RESULTADOS,
        "predicciones_octavos_2026.csv"
    )
)

print("Archivos de resultados cargados correctamente.")
```

## 2. Comparación de modelos

Para evaluar los algoritmos se utilizan las métricas Accuracy, Precision, Recall y F1-score.

La comparación permite identificar el modelo con mejor equilibrio entre capacidad general de clasificación y detección correcta de los equipos clasificados.

```python
comparacion_modelos
```

```python
metricas = [
    "Accuracy",
    "Precision",
    "Recall",
    "F1-score"
]

ax = comparacion_modelos.set_index(
    "Modelo"
)[metricas].plot(
    kind="bar",
    figsize=(14, 7)
)

ax.set_title(
    "Comparación del rendimiento de los modelos de Machine Learning"
)

ax.set_xlabel("Modelo")
ax.set_ylabel("Valor de la métrica")
ax.set_ylim(0, 1)

plt.xticks(
    rotation=30,
    ha="right"
)

plt.tight_layout()
plt.show()
```

## 3. Modelo seleccionado

De acuerdo con la comparación experimental, SVM fue seleccionado como modelo final.

El modelo alcanzó un Accuracy de 0.7333 y un F1-score de 0.8333 durante la comparación principal.

La selección considera especialmente el F1-score, debido a que esta métrica combina Precision y Recall y permite evaluar de forma equilibrada la capacidad de clasificación.

## 4. Validación temporal con el Mundial 2022

Para evitar evaluar el modelo con información utilizada durante el entrenamiento, se aplicó una división temporal.

Los Mundiales comprendidos entre 2002 y 2018 fueron utilizados como información histórica de entrenamiento.

El Mundial 2022 se utilizó como prueba independiente.

Este procedimiento permite analizar la capacidad del modelo para generalizar hacia un torneo posterior.

```python
validacion_2022
```

```python
total_partidos = len(validacion_2022)

aciertos = (
    validacion_2022["resultado_prediccion"]
    == "CORRECTO"
).sum()

porcentaje_acierto = (
    aciertos / total_partidos
) * 100

print("Partidos evaluados:", total_partidos)
print("Predicciones correctas:", aciertos)
print(
    f"Porcentaje de acierto: "
    f"{porcentaje_acierto:.2f}%"
)
```

La matriz de confusión permite observar la cantidad de clasificaciones correctas e incorrectas realizadas por el modelo.

## 6. Curva ROC y AUC

La curva ROC representa la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos.

El AUC resume la capacidad discriminativa del modelo. Un valor superior a 0.5 indica un comportamiento superior a una clasificación aleatoria.
